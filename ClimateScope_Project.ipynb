```python
import pandas as pd

df = pd.read_csv('GlobalWeatherRepository.csv')

print("Dataset Info:")
df.info()
```

* The dataset contains 89,248 entries and 41 columns.
* It has a mix of data types, including object (e.g., country, location_name,
condition_text), int64 (e.g., humidity, cloud, wind_degree), and float64 (e.g., temperature_celsius, wind_kph, uv_index).


```python
print("\nFirst 5 rows of the dataset:")
df.head()
```

```python
df.isnull()
# Dropping rows with any missing data for simplicity. A more complex project might use imputation.
df.dropna(inplace=True)
print(f"\nDataFrame size after dropping missing values: {df.shape}")
```

* It shows no missing values, as all columns have 89,248 non-null entries.



```python
# Identify anomalies by checking descriptive statistics for numerical columns
print("Descriptive statistics for numerical columns:")
df.describe()
```

```python
# Check for inconsistencies in key categorical columns
print("\nUnique values for 'condition_text':")
print(df['condition_text'].unique())

print("\nUnique values for 'wind_direction':")
print(df['wind_direction'].unique())

print("\nUnique values for 'timezone':")
print(df['timezone'].unique())
```

* The unique values for condition_text show some inconsistencies in capitalization (e.g., 'Partly Cloudy' and 'Partly cloudy'). While this doesn't impact numerical analysis, it should be addressed for consistent categorization and visualization. The wind_direction and timezone columns appear to be clean, with no obvious inconsistencies.

```python
# Determine the data coverage (time range)
df['last_updated'] = pd.to_datetime(df['last_updated'])
min_date = df['last_updated'].min()
max_date = df['last_updated'].max()
print(f"\nData coverage from {min_date} to {max_date}")
```

```python
# Let's convert 'last_updated' to a datetime object for easier aggregation.
df['last_updated'] = pd.to_datetime(df['last_updated'])
df['month'] = df['last_updated'].dt.month
df['year'] = df['last_updated'].dt.year

# This aggregates key metrics by month and country.
monthly_avg_df = df.groupby(['country', 'year', 'month']).agg({
    'temperature_celsius': 'mean',
    'humidity': 'mean',
    'wind_kph': 'mean',
    'precip_mm': 'sum'
}).reset_index()

print("\nMonthly aggregated data (first 5 rows):")
monthly_avg_df.head()
```

```python
# Task: Perform statistical analysis to understand distributions, correlations, etc.
print("Statistical Summary of Numerical Columns:")
df.describe()
```

```python
# Find the top 5 hottest and coldest records
hottest_records = df.nlargest(5, 'temperature_celsius')[['country', 'temperature_celsius', 'last_updated', 'condition_text']]
coldest_records = df.nsmallest(5, 'temperature_celsius')[['country', 'temperature_celsius', 'last_updated', 'condition_text']]

print("\nTop 5 Hottest Weather Records:")
print(hottest_records)

print("\nTop 5 Coldest Weather Records:")
print(coldest_records)
```

```python
# Let's compare the average temperature and humidity for a few random countries.
top_countries = df['country'].value_counts().head(5).index.tolist()
regional_avg = df[df['country'].isin(top_countries)].groupby('country').agg({
    'temperature_celsius': 'mean',
    'humidity': 'mean'
}).reset_index()

print("\nComparison of average temperature and humidity for top 5 represented countries:")
print(regional_avg)
```

```python
import plotly.express as px

# Let's ensure the data is in a format suitable for plotting
avg_temp_country = df.groupby('country').agg({'temperature_celsius': 'mean', 'humidity': 'mean'}).reset_index()

# Task: Build static and interactive visualizations
print("\nChoropleth Map...")
fig_map = px.choropleth(
    avg_temp_country,
    locations="country",
    locationmode='country names',
    color="temperature_celsius",
    hover_name="country",
    hover_data={'temperature_celsius': ':.2f', 'humidity': True},
    color_continuous_scale=px.colors.sequential.Plasma,
    title="Average Temperature per Country"
)
fig_map.show()
```

```python
# Aggregate data by month for the line chart
df['month_year'] = df['last_updated'].dt.to_period('M').astype(str)
monthly_trend = df.groupby('month_year').agg({
    'temperature_celsius': 'mean'
}).reset_index()

print("\nLine Chart for Temperature Trend...")
fig_line = px.line(
    monthly_trend,
    x="month_year",
    y="temperature_celsius",
    title="Monthly Average Temperature Trend",
    labels={'temperature_celsius': 'Avg Temperature (°C)', 'month_year': 'Date'}
)
fig_line.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

#heatmap for correlation
print("\nCorrelation Heatmap...")
# Select numerical columns for correlation
numerical_cols = ['temperature_celsius', 'humidity', 'pressure_mb', 'wind_kph', 'uv_index']
corr_matrix = df[numerical_cols].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Heatmap of Weather Variables')
plt.show()
```

```python
print("\n--- Average Wind Speed per Country ---")
# Group by country to calculate the mean wind speed
avg_wind_country = df.groupby('country').agg({
    'wind_kph': 'mean'
}).reset_index()

# Sort for better visualization and select top 10
top_10_windy = avg_wind_country.nlargest(10, 'wind_kph')

print("Bar Chart for Top 10 Countries by Wind Speed...")
fig_bar_wind = px.bar(
    top_10_windy,
    x='country',
    y='wind_kph',
    title='Top 10 Countries by Average Wind Speed (km/h)',
    labels={'wind_kph': 'Average Wind Speed (km/h)', 'country': 'Country'}
)
fig_bar_wind.show()
```

```python
#Summarize regional and global climate trends
global_avg_temp = df['temperature_celsius'].mean()
global_max_temp = df['temperature_celsius'].max()
global_min_temp = df['temperature_celsius'].min()

print("\nGlobal Trends Summary:")
print(f"Global Average Temperature: {global_avg_temp:.2f}°C")
print(f"Global Maximum Temperature Recorded: {global_max_temp:.2f}°C")
print(f"Global Minimum Temperature Recorded: {global_min_temp:.2f}°C")
```
